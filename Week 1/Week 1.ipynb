Requirements:

-Rename the new field with the Bank code 'Bank'. 

-Rename the values in the Online or In-person field, Online of the 1 values and In-Person for the 2 values. 

-Change the date to be the day of the week

-Different levels of detail are required in the outputs. You will need to sum up the values of the transactions in three ways:
1. Total Values of Transactions by each bank
2. Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)
3. Total Values by Bank and Customer Code

-Output each data file 

```python
df = pd.read_csv("PD 2023 Wk 1 Input.csv")
df.head()
```

```python
#Split Transaction code to extract letters at the start to get bank and transaction code
df[['Bank', 'Transaction Code']] = df['Transaction Code'].str.split('-',n=1, expand = True)
```

```python
#Replace 1 with Online and 2 with In-Person
df['Online or In-Person'] = df['Online or In-Person'].replace({1: 'Online', 2: 'In-Person'})
```

```python
#Change Transaction Date to day of the week
df['Transaction Date'] = df['Transaction Date'].astype('datetime64[D]')
df['Transaction Date'] = df['Transaction Date'].dt.day_name()
```

```python
#Total Values of Transactions by each bank
df_pivot1 = df.pivot_table(index=['Bank'],values=['Value'], aggfunc='sum' )
df_pivot1
```

```python
#Total Values by Bank, Day of the Week and Type of Transaction (Online or In-Person)
df_pivot2 = df.pivot_table(index=['Bank','Transaction Date','Online or In-Person'],values=['Value'], aggfunc='sum' )
df_pivot2
```

```python
#Total Values by Bank and Customer Code
df_pivot3 = df.pivot_table(index=['Bank','Customer Code'],values=['Value'], aggfunc='sum' )
df_pivot3
```
