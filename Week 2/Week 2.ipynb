Requirements
- Input the data

- In the Transactions table, there is a Sort Code field which contains dashes. We need to remove these so just have a 6 digit string 

- Use the SWIFT Bank Code lookup table to bring in additional information about the SWIFT code and Check Digits of the receiving bank account 

- Add a field for the Country Code

Hint: all these transactions take place in the UK so the Country Code should be GB
- Create the IBAN as above

```python
#Remove the dashes in the Sort Code
trans['Sort Code'] = trans['Sort Code'].str.replace('-','')
trans
```

```python
#Use the SWIFT Bank Code lookup table to bring in additional information about the SWIFT code and Check Digits of the receiving bank account 

df = pd.merge(trans, swifts, on='Bank')
df
```

```python
#Add a field for country code, in our case, all should be GB
df['Country Code'] = 'GB'
df
```

```python
#Create IBAN (Country Code + Check Digits + Bank Code + Sort Code + Account Number)
df['IBAN'] = df['Country Code'] + df['Check Digits'].astype(str) + df['SWIFT code'].astype(str) + df['Sort Code'].astype(str) + df["Account Number"].astype(str)


df
```

```python
#Keep Only Transaction ID and IBAN

df = df[['Transaction ID', 'IBAN']]
```
