```python
import pandas as pd
```

Requirements
- Input data
- Create the bank code by splitting out off the letters from the Transaction code, call this field 'Bank'
- Change transaction date to the just be the month of the transaction
- Total up the transaction values so you have one row for each bank and month combination
- Rank each bank for their value of transactions each month against the other banks. 1st is the highest value of transactions, 3rd the lowest. 
- Without losing all of the other data fields, find:
    - The average rank a bank has across all of the months, call this field 'Avg Rank per Bank'
    - The average transaction value per rank, call this field 'Avg Transaction Value per Rank'
    
Output the data

```python
df = pd.read_csv('PD 2023 Wk 1 Input.csv')
```

```python
#Split bank code
df[['Bank','Transaction Code']] = df['Transaction Code'].str.split('-',n=1,expand= True)
```

```python
#Change Transaction Date to Month
df['Transaction Date'] = df['Transaction Date'].astype('datetime64[D]')
df['Month'] = df['Transaction Date'].dt.month_name()
```

```python
#Aggragate the sum of Value by Bank and Month
df_pivot = ( pd.pivot_table(df,index=['Bank', 'Month'], values='Value', aggfunc='sum').reset_index())
```

```python
#Sort table and Rank and the banks for each month
df2 = df_pivot.sort_values(['Month', 'Value'],ascending = [True, False])
df2['Rank'] = df2.groupby('Month').cumcount() +1
```

```python
#Calculate Average Rank for each bank
average_rank = df2.groupby('Bank')['Rank'].mean().reset_index()
average_rank.rename(columns={'Rank': 'Average Rank'}, inplace=True)
df2 = pd.merge(df2, average_rank, on='Bank')
```

```python
#Calculate Average value for each rank
average_value_per_rank = df2.groupby('Rank')['Value'].mean().reset_index()
average_value_per_rank.rename(columns={'Value': 'Average Value Per Rank'}, inplace=True)
df2 = pd.merge(df2, average_value_per_rank, on='Rank')
```
